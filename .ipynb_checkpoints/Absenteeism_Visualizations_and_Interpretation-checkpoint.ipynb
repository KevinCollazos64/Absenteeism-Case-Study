## Reason vs. Probability

```python
Image(filename='Reason_vs_Probability.png')
```

### Removed reason 2 because in none of the 40 new observations in the new dataset was there a case where someone was excessively absent from work due to reason 2, and was verified in the data source.
### If a person is absent due to reason 1, the probability that they will be excessively absent is > 50% (so they are expected to be away from work for more than 3 hours).
### On the contrary, reason 3 has few observations where people specify this reason for being excessively absent. The probabilities are disributed in both the upper and lower parts of the 50% probability line. Therefore, similarly to reason 2, we cannot deduce much about what to expect from people who are absent because of reason 3.
### In reason 4, we see that people who attributed their absence to this reason group have a < 50 % probability to be excessively absent.
### Referring back to the reasons chart, we see that reasons in group 1 are very serious illnesses. It makes sense that the expected probability for someone to be excessively absent due to reason 1 is over 50%.
### Additionally, the results for reason 4 are logical because reason 4 represents non-life threatening reasons, such as a dental appointment, physical therapy and medical consultations. It is reasonable to assume that a person that is absent for this reason will not be gone from the workplace for an entire day. This is why the plot shows that people who are absent because of reason 4 are not projected to be excessively absent with a predicted probability of sub 50%.
